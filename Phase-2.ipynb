```python
# import libraries
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Display all columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
df['OverallSatisfaction'] = ( df['EnvironmentSatisfaction'] + df['JobSatisfaction'] + df['RelationshipSatisfaction'] + df['WorkLifeBalance'] ) / 4
df['TenureRatio'] = ( df['YearsAtCompany'] / df['TotalWorkingYears'].replace(0, np.nan).fillna(0) )
```

```python
df = df.drop(columns=[
    "EnvironmentSatisfaction","JobSatisfaction","RelationshipSatisfaction","WorkLifeBalance","YearsAtCompany","TotalWorkingYears",
    "Over18","EmployeeCount","EmployeeNumber","StandardHours","DailyRate","EducationField","MonthlyIncome","MonthlyRate","TrainingTimesLastYear"])
```

```python
df.head()
```

```python
df['OverTime'] = df['OverTime'].map({ 'Yes':1, 'No':0 })
df['Attrition'] = df['Attrition'].map({ 'Yes':1, 'No':0 })
df['BusinessTravel'] = df['BusinessTravel'].map({ 'Travel_Frequently':2, 'Travel_Rarely':1, 'Non-Travel':0 })
df['Gender'] = df['Gender'].map({ 'Female':1, 'Male':0 })
```

```python
df['Department'] = df['Department'].map({'Research & Development':2, 'Sales':1, 'Human Resources':0 })
```

```python
df['JobRole'] = df['JobRole'].map({'Healthcare Representative': 0,'Human Resources': 1,'Laboratory Technician': 2,'Manager': 3,
'Manufacturing Director': 4,'Research Director': 5,'Research Scientist': 6,'Sales Executive': 7,'Sales Representative': 8})
```

```python
df['MaritalStatus'] = df['MaritalStatus'].map({'Single': 0,'Married': 1,'Divorced': 2})
```
