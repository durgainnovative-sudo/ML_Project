```python
#Removing unnecessary columns
df.drop(
    columns=[
        "EmployeeCount",
        "EmployeeNumber",
        "Over18",
        "StandardHours"
    ],
    inplace=True
)
```

```python
df.columns
```

```python
#seperating numerical and categorical columns
categorical_cols = df.select_dtypes(include='object').columns

numerical_cols = df.select_dtypes(exclude='object').columns

print(categorical_cols)
print(numerical_cols)
```

```python
#Target variable
df['Attrition'].value_counts()
```

```python
df['Attrition'].value_counts(normalize=True)*100
```

```python
#EDA
#Attrition distribution
plt.figure(figsize=(6,5))

sns.countplot(x='Attrition', data=df)

plt.title("Employee Attrition Distribution")

plt.show()
```

```python
#Age distribution
plt.figure(figsize=(8,5))

sns.histplot(df['Age'], bins=20, kde=True)

plt.title("Age Distribution")

plt.show()
```

```python
#Attrition by Gender
plt.figure(figsize=(6,5))

sns.countplot(
    x='Gender',
    hue='Attrition',
    data=df
)

plt.title("Attrition by Gender")

plt.show()
```

```python
#Attrition by Department
plt.figure(figsize=(8,5))

sns.countplot(
    x='Department',
    hue='Attrition',
    data=df
)

plt.xticks(rotation=20)

plt.show()
```

```python
#Attrition by Role
plt.figure(figsize=(12,5))

sns.countplot(
    y='JobRole',
    hue='Attrition',
    data=df
)

plt.show()
```

```python
#Monthly Income Distribution
plt.figure(figsize=(8,5))

sns.histplot(df['MonthlyIncome'], bins=30)

plt.show()
```

```python
#Monthly Income vs Attrition
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Attrition',
    y='MonthlyIncome',
    data=df
)

plt.show()
```

```python
#Overtime vs Attrition
plt.figure(figsize=(6,5))

sns.countplot(
    x='OverTime',
    hue='Attrition',
    data=df
)

plt.show()
```

```python
#Job satisfaction vs Atrrition
plt.figure(figsize=(8,5))

sns.countplot(
    x='JobSatisfaction',
    hue='Attrition',
    data=df
)

plt.show()
```

```python
#Worklife balance vs Attrition
plt.figure(figsize=(8,5))

sns.countplot(
    x='WorkLifeBalance',
    hue='Attrition',
    data=df
)

plt.show()
```

```python
#Correlation Matrix
plt.figure(figsize=(15,12))

corr = df.select_dtypes(include=np.number).corr()

sns.heatmap(
    corr,
    cmap='coolwarm',
    annot=True,
    fmt=".2f"
)

plt.show()
```

```python
#Pair plot
sns.pairplot(
    df[
        [
            'Age',
            'MonthlyIncome',
            'YearsAtCompany',
            'TotalWorkingYears',
            'Attrition'
        ]
    ],
    hue='Attrition'
)

plt.show()
```
