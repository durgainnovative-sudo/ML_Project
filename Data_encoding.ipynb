```python
# import libraries
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Display all columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

## Dataset with "OverallSatisfaction" & "TenureRatio" with encoding

```python
df['OverallSatisfaction'] = ( df['EnvironmentSatisfaction'] + df['JobSatisfaction'] + df['RelationshipSatisfaction'] + df['WorkLifeBalance'] ) / 4
df['TenureRatio'] = ( df['YearsAtCompany'] / df['TotalWorkingYears'].replace(0, np.nan).fillna(0) )
```

```python
df = df.drop(columns=[
    "EnvironmentSatisfaction","JobSatisfaction","RelationshipSatisfaction","WorkLifeBalance","YearsAtCompany","TotalWorkingYears",
    "Over18","EmployeeCount","EmployeeNumber","StandardHours","DailyRate","EducationField","MonthlyIncome","MonthlyRate","TrainingTimesLastYear"])
```

```python
df['OverTime'] = df['OverTime'].map({ 'Yes':1, 'No':0 })
df['Attrition'] = df['Attrition'].map({ 'Yes':0, 'No':1 })
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_columns = ["BusinessTravel","Gender","Department","JobRole","MaritalStatus"]

X_cat = df[categorical_columns]



encoder = OneHotEncoder(drop='first').fit(X_cat)
cols = encoder.get_feature_names_out(input_features=X_cat.columns)
X_cat_encode = pd.DataFrame(encoder.transform(X_cat).toarray(),columns=cols)
X_cat_encode.head()
```

```python
df_encode = pd.concat([df, X_cat_encode], axis=1)
#df_encode

#df_encode = df_encode.drop(columns=["BusinessTravel","Gender","Department","JobRole","MaritalStatus"])
df_encode=df_encode.fillna(0)
```
